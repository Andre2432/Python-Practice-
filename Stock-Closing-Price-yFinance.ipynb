```python
import pandas as pd # Dataframe Library
import matplotlib.pyplot as plt # Plot Chart Library
```

```python
import yfinance as yf
import datetime as dt 
import mplfinance as mpf

yf.pdr_override()

# Stock input
aapl = yf.Ticker('AAPL')
amzn = yf.Ticker('AMZN')

#Date
start = dt.datetime(2022, 11, 1)
end = dt.datetime.now()
```

```python
#@title
aapl.info
```

```python
#@title
amzn.info
```

```python
# dataframe
data = yf.download('AAPL', start, end)
data1 = yf.download('AMZN', start, end)
```

```python
#AAPL CLOSING PRICE GRAPH PLOT LINE GRAPH
print('AAPL')
plt.figure(figsize=(14,8))
plt.scatter(data.index, data['Adj Close'], color='black')
plt.plot(data.index, data['Adj Close'], color='r')
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()

#AMZN CLOSING PRICE PLOT LINE GRAPH
print('AMZN')
plt.figure(figsize=(14,8))
plt.scatter(data.index, data1['Adj Close'], color='black')
plt.plot(data.index, data1['Adj Close'], color='r')
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```

```python
#AAPL CLOSING PRICE CANDLE GRAPH
print('AAPL')
#mpf.plot(data,type='candle' ,mav=(3,6,9))
mpf.plot(data.iloc[0:40],type='candle',mav=(3,6,9),update_width_config=dict(candle_linewidth=0.4,candle_width=1.6))


#AMZN CLOSING PRICE CANDLE GRAPH
print('AMZN')
mpf.plot(data1.iloc[0:40],type='candle',mav=(3,6,9),update_width_config=dict(candle_linewidth=0.4,candle_width=1.6))
```

```python
# dataframe
data = yf.download('AAPL',start,end)

data['VolumePositive'] = data['Open'] < data['Adj Close']
#CLOSING PRICE LINE CHART 
print('Line Chart')
fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(3, 1, 1)
ax1.plot(data['Adj Close'])
ax1.set_title('Stock '+ 'AAPL' +' Closing Price')
ax1.set_ylabel('Price')
ax1.legend(loc='best')

#CLOSING PRICE VOLUME LINE CHART
ax2 = plt.subplot(3, 1, 2)
ax2.plot(data['Volume'], label='Volume')
ax2.grid()
ax2.legend(loc='best')
ax2.set_ylabel('Volume')

#CLOSING PRICE VOLUME BAR CHART
ax3 = plt.subplot(3, 1, 3)
ax3v = ax3.twinx()
colors = data.VolumePositive.map({True: 'g', False: 'r'})
ax3v.bar(data.index, data['Volume'], color=colors, alpha=0.4)
ax3.set_ylabel('Volume')
ax3.grid()
ax3.set_xlabel('Date')

# dataframe
data = yf.download('AMZN',start,end)

data['VolumePositive'] = data['Open'] < data['Adj Close']
#CLOSING PRICE LINE CHART 
print('Line Chart')
fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(3, 1, 1)
ax1.plot(data['Adj Close'])
ax1.set_title('Stock '+ 'AMZN' +' Closing Price')
ax1.set_ylabel('Price')
ax1.legend(loc='best')

#CLOSING PRICE VOLUME LINE CHART
ax2 = plt.subplot(3, 1, 2)
ax2.plot(data['Volume'], label='Volume')
ax2.grid()
ax2.legend(loc='best')
ax2.set_ylabel('Volume')

#CLOSING PRICE VOLUME BAR CHART
ax3 = plt.subplot(3, 1, 3)
ax3v = ax3.twinx()
colors = data.VolumePositive.map({True: 'g', False: 'r'})
ax3v.bar(data.index, data['Volume'], color=colors, alpha=0.4)
ax3.set_ylabel('Volume')
ax3.grid()
ax3.set_xlabel('Date')




```
